```python
%matplotlib inline
from matplotlib import pylab as plt
from python_dist import NormalDist
```

```python
a = NormalDist(10, 5)
b = NormalDist(20, 2)
```

```python
plt.hist(a.get_samples(100000), bins=100)
plt.hist(b.get_samples(100000), bins=100)
plt.show()
```

```python
c = a + b
plt.hist(c.get_samples(100000), bins=100)
plt.show()
```
